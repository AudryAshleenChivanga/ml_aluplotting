```python
# Bin size 20
plt.figure(figsize=(10, 6))
plt.hist(df['co2'], bins=20, color='skyblue', edgecolor='black')
plt.title('Carbon Dioxide Emissions Distribution (20 Bins)')
plt.xlabel('Carbon Dioxide Emissions')
plt.ylabel('Frequency')
plt.savefig("CO2emissionsdistributions.png")
plt.show()
```

```python
# Bin size 40
plt.figure(figsize=(10, 6))
plt.hist(df['co2'], bins=40, color='skyblue', edgecolor='black')
plt.title('Carbon Dioxide Emissions Distribution (40 Bins)')
plt.xlabel('Carbon Dioxide Emissions')
plt.ylabel('Frequency')
plt.savefig("CO2emissions.png")
plt.show()
```

```python
# Histogram for City Fuel Economy
plt.figure(figsize=(10, 6))
plt.hist(df['city'], bins=20, color='orange', edgecolor='black')
plt.title('City Fuel Economy Distribution')
plt.xlabel('City Fuel Economy (mpg)')
plt.savefig("CityFuel.png")
plt.ylabel('Frequency')
plt.show()
```

```python
# Histogram for Greenhouse Gas Score
plt.figure(figsize=(10, 6))
plt.hist(df['ghgScore'], bins=10, color='teal', edgecolor='black')
plt.title('Distribution of Greenhouse Gas Scores')
plt.xlabel('Greenhouse Gas Score')
plt.savefig("greenhousegasscore.png")
plt.ylabel('Frequency')
plt.show()
```

```python
correlation_matrix = df.corr()
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, center=0, vmin=-1, vmax=1, cmap='vlag')
plt.title('Correlation Heat Map')
plt.savefig("Heatmap.png")

plt.show()
```

```python
### Heatmap Interpretation

The heatmap highlights significant relationships between various features in the dataset, particularly those associated with fuel efficiency and environmental impact.

1. **Strong Positive Correlations:**
   - Features such as `city`, `UCity`, `highway`, `UHighway`, and `comb` demonstrate strong positive correlations with each other. These metrics are directly linked and collectively measure different aspects of fuel efficiency. This indicates that improvements in one metric are likely to be reflected in the others, making these features reliable indicators of a vehicle's overall fuel economy.

2. **Negative Correlation with CO₂ Emissions:**
   - CO₂ emissions (`co2`) show a strong negative correlation with both the Greenhouse Gas Score (`ghgScore`) and the Fuel Economy Score (`feScore`). This suggests that as CO₂ emissions increase, environmental scores decrease, indicating a greater negative environmental impact. Therefore, vehicles with lower CO₂ emissions tend to have higher `ghgScore` and `feScore` values, signifying better environmental performance.

3. **Feature Selection for Predictive Modeling:**
   - If the goal is to predict the `ghgScore`, some features might be less useful. For example, `pv2` and `pv4` have low correlations with `ghgScore`, meaning they do not significantly contribute to predicting this target and could be excluded. Additionally, categorical features like `make`, `model`, `VClass`, `drive`, `trans`, and `fuelType` may not directly correlate with `ghgScore` in their raw form, but they could become valuable predictors if appropriately encoded.

4. **Multicollinearity Consideration:**
   - The `feScore` and `ghgScore` are highly correlated, so if `ghgScore` is the target variable, `feScore` might be redundant. Removing `feScore` could help avoid multicollinearity issues in predictive modeling.
```
